```python
from bing_image_downloader import downloader
```

```python
downloader.download('car',limit=30,output_dir='bike_vs_car',adult_filter_off=True)
```

```python
downloader.download('bike',limit=30,output_dir='bike_vs_car',adult_filter_off=True)
```

```python
datadir='bike_vs_car'
img_size=100
categories=['bike','car']
import os
import cv2
import numpy as np
import random
import matplotlib.pyplot as pp
data=[]
for category in categories:
    folder_path=os.path.join(datadir,category)
    label=categories.index(category)
    for img in os.listdir(folder_path):
        img_path=os.path.join(folder_path,img)
        img_read=cv2.imread(img_path)
        img_read=cv2.resize(img_read,(img_size,img_size))
        data.append([img_read,label])
        #pp.imshow(img_read)
        
        
```

```python
len(data)
```

```python
random.shuffle(data)
```

```python
x=[]
y=[]
for features,labels in data:
    x.append(features)
    y.append(labels)
```

```python
x=np.array(x)
y=np.array(y)
```
