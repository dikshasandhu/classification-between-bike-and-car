```python
import pickle
features=pickle.load(open('bike_vs_car.pkl','rb'))
labels=pickle.load(open('labels_b_c.pkl','rb'))
```

```python
features[0]
```

```python
features=features/255.0
```

```python
features[0]
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Dense,Flatten
model=Sequential()
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Flatten())
model.add(Dense(64,input_shape=features.shape[1:],activation='relu'))
model.add(Dense(2,activation='softmax'))
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
model.fit(features,labels,validation_split=0.2,epochs=40)
```
